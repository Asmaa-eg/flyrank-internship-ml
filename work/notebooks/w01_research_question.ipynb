## 1. My lane (or freestyle) and why

I’m choosing the Refresh / Content Opportunity lane. This feels like the best fit because the starter data contains observable signals about content age, visibility, and trend, and the goal is to help a team decide which pages are worth reviewing first. I’m especially interested in whether a simple ranking approach can surface pages that look stale but still have enough demand to matter.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import os
import subprocess
import sys
from pathlib import Path

import pandas as pd

REPO_URL = "https://github.com/flyrank-bih/flyrank-ml-internship-starter"
REPO_DIR = "flyrank-ml-internship-starter"


def find_csv():
    roots = [Path.cwd(), *Path.cwd().parents]
    roots += [Path("/content"), Path("/content/" + REPO_DIR), Path("/mnt/data")]
    for base in roots:
        candidate = base / "data" / "raw" / "content_refresh_anonymized.csv"
        if candidate.exists():
            return candidate
    return None


csv_path = find_csv()
if csv_path is None:
    if "google.colab" in sys.modules:
        if not os.path.isdir(REPO_DIR):
            subprocess.run(["git", "clone", "--depth", "1", REPO_URL, REPO_DIR], check=True)
        os.chdir(REPO_DIR)
        csv_path = find_csv()

if csv_path is None:
    raise FileNotFoundError("Could not find data/raw/content_refresh_anonymized.csv. Please make sure the repo is available in this environment.")


df = pd.read_csv(csv_path)
print("Rows:", len(df))
print("Declining rate:", round((df["trend_direction"].str.lower() == "down").mean(), 3))
print("Pages with impressions > 0:", int((df["impressions_90d"] > 0).sum()))
print("Average impressions:", round(df["impressions_90d"].mean(), 2))
```

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
print("Pages with impressions >= 500:", int((df["impressions_90d"] >= 500).sum()))
print("Pages with age >= 180 days:", int((df["content_age_days"] >= 180).sum()))
print("Pages with both conditions:", int(((df["impressions_90d"] >= 500) & (df["content_age_days"] >= 180)).sum()))
```
